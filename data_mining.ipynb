```python
def create_collection_dict(paths, titles, author):
    """Function for turning each txt book file into a json file for modeling
        
    Args:
        paths (list): All the paths of the text files with full book text
        titles (list): Titles of the books
        author (str): Author of the book
    
    Returns:
        dict_collection (dict): Dictionary of metadata and text prepared for modeling. Each element 
                                 of the value for the "text" key should be a stirng with an entire 
                                 chapter/act of the book
    
    """
        
    dict_collection = []
    path_count = 0
    
    for book_path in paths:
        
        book_txt = open(book_path, 'r')
        book_txt = book_txt.read()
        
        act_dict = {}
        is_act = False

        if "ACT " in book_txt:
            is_act = True
            act_list = re.split("ACT ", book_txt)

        elif "Actus " in book_txt:
            is_act = True
            act_list = re.split("Actus ", book_txt)

        else:
            print("ACT nor Actus was found")
            break

        del act_list[0]

        for i, act in enumerate(act_list):
            temp_dict = {}
            temp_dict["Title"] = titles[path_count]
            temp_dict["Author"] = author

            if is_act:
                temp_dict["Act"] = str(i+1)
            else:
                temp_dict["Chapter"] = str(i+1)

            temp_dict["Text"] = str(act)

            dict_collection.append(temp_dict)
            
        path_count += 1
        
    return dict_collection
```
